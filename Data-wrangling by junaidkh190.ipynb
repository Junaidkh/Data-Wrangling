```
# Taking input csv files 
from google.colab import files
uploaded = files.upload()
```

```
# Importing Libraries
import pandas as pd
import io
from datetime import datetime
import calendar
```

```
# Reading csv file into dataframe
df = pd.read_csv(io.StringIO(uploaded['ipl_2019_matches - Source.csv'].decode('utf-8')))
```

```
df.head(5)
```

```
# creating second empty dataframe
df2 = pd.DataFrame()
```

```

# storing day, month, year in dataframe
df2['year'] = pd.DatetimeIndex(df['Date']).year
df2['month'] = pd.DatetimeIndex(df['Date']).month
df2['day'] = pd.DatetimeIndex(df['Date']).day
```

```
df2.head()
```

```
# storing date into dataframe
df['Date'] = pd.to_datetime(df['Date'])
```

```
# Getting day of week
df2['DayOfWeek']= df['Date'].dt.dayofweek
```

```
# Getting weekday name
df2['day'] = df['Date'].dt.weekday_name
```

```
df2.head()
```

```
# Finding whether the day is weekday or Weekend
df2['Week_Type'] = 'Weekday'          # Initialize the column with default value of 'WeekDay'
df2.loc[df2['DayOfWeek'].isin([5, 6]), 'Week_Type'] = 'Weekend'
```

```
# Getting name of the month
df2['Month_name'] = df['Date'].dt.strftime('%B')
```

```

df2['Date'] = df['Date'].dt.date
```

```
# Getting Match number 
df2['Match_No'] = df.iloc[:,0]
```

```
# Getting Match played between teams
df2['Match'] = df['Match']
```

```
df2.head()
```

```
# Finding out which two teams have played
df2[['Team_1','Team_2']] = df2['Match'].apply(lambda x: pd.Series(str(x).split('v')))
```

```
df2['Team_1'] = df2['Team_1'].apply(lambda x: pd.Series(str(x[11:])))
```

```
# Getting ground location on which cricket has been played
df2['Location'] = df['Ground']
```

```
# Finding out the result
df2['Result'] = df['Result']
```

```
# Finding out which team has won the match
df2['Winning_Team'] = df2['Result'].apply(lambda x: pd.Series(str(x).split('won')[0]))
```

```
# Mode of winning the match
mode=[]
for i in df2.Result:
  if i.split(' ')[-1]=='Runs':
    mode.append('Runs')
  elif i.split(' ')[-1]=='Wickets':
    mode.append('Wickets')
  else:
    mode.append(i)

df2['Winning_mode'] = mode
```

```
# By how much margin match has been won
margin=[]
for i in df2.Result:
  if i=='No Result':
    margin.append('No Result')
  elif i.split()[-1]=='eliminator)':
    margin.append(i)
  else:
    margin.append(i.split()[-2])
df2['Winning_Margin']=margin
```

```
# Finding out which team bats first
df2['Team_Batting_First'] = df['Match'].apply(lambda x: pd.Series(str(x).split('v')[1]))
```

```
# Rearranging Coulmns
df2 = df2[['Match_No','Date','year','month','Month_name','day','DayOfWeek','Week_Type','Match','Team_1','Team_2','Location','Result','Winning_Team','Winning_mode','Winning_Margin','Team_Batting_First']]
```
